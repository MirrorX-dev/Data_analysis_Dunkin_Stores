```python
datos_establecimientos  = pd.read_csv('C:/Users/Vespertino/data/dunkin_stores.csv')
datos_establecimientos.dropna(inplace=True)
datos_establecimientos .head(10)
```

```python
#Importar datos de ciudades americanas:
datos_ciudades = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/us-cities-top-1k.csv")
datos_ciudades.head()
```

```python
#Muestra en un mapa los datos, campos lat, long.
#fig = px.scatter_geo(datos, lat = "loc_lat", lon = "loc_long", scope = 'usa', center = {"lat": 40, "lon": 0}, fitbounds ='locations', opacity = 0.1)
fig = px.scatter_mapbox(datos_establecimientos , lat="loc_lat", lon="loc_long", zoom=3)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```

```python
# Agregar marcadores de ciudades al mapa
fig.add_trace(px.scatter_mapbox(datos_ciudades, lat="lat", lon="lon", 
                                 hover_name="City", hover_data=["State", "Population"], 
                                 color_discrete_sequence=["fuchsia"], zoom=3).data[0])

# Configurar el dise√±o del mapa
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

# Mostrar el mapa
fig.show()
```

```python
estados_agrupados = datos_ciudades.groupby('State').agg({'City': 'count', 'Population': 'sum'})
df_estado_info = pd.DataFrame(estados_agrupados)


# Mostrar el dataframe
print(estados_agrupados)
```

```python
#N¬∫ de ciudades y poblaci√≥n total en cada estado de los EEUU
# Agrupar por estado y calcular la cantidad de ciudades y la poblaci√≥n total
datos_agrupados = datos_ciudades.groupby('State').agg({'City': 'count', 'Population': 'sum'})

# Cambiar los nombres de las columnas del dataframe resultante
datos_agrupados = datos_agrupados.rename(columns={'City': 'N¬∫ de Ciudades', 'Population': 'Poblaci√≥n Total'})

# Mostrar el dataframe resultante
print(datos_agrupados)
```

```python
#N¬∫ de ciudades por provincia en Espa√±a tomados de un CSV de c√≥dgios postales:
datos_spain = pd.read_csv("../datos/listado-codigos-postales-con-LatyLon.csv", sep = ";")
datos_spain.head()
```

```python
#Crear dataframe:

# Eliminar filas duplicadas
datos_spain = datos_spain.drop_duplicates(subset=['provincia', 'poblacion', 'codigopostalid'])

# Agrupar los datos por provincia
datos_por_provincia = datos_spain.groupby('provincia')

# Contar el n√∫mero de ciudades por provincia
num_ciudades_por_provincia = datos_por_provincia['poblacion'].nunique()

# Mostrar el resultado
print(num_ciudades_por_provincia)
```
